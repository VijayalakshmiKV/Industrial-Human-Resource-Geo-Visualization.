```python
# read & merge all CSV files
# List to store individual DataFrames   
df_list = []

# Loop through each file in the folder
for file in os.listdir(DATA_PATH):
    
    # Read only CSV files (handles .csv and .CSV)
    if file.lower().endswith(".csv"):
        file_path = os.path.join(DATA_PATH, file)
        
        # Read CSV with proper encoding (important for census data)
        df = pd.read_csv(file_path, encoding="latin1")
        
        # Add source file column for state identification
        df["source_file"] = file
        
        # Append DataFrame to list
        df_list.append(df)

# Merge all DataFrames into one
merged_df = pd.concat(df_list, ignore_index=True)

# Display dataset shape
print("Number of files merged:", len(df_list))
print("Total rows and columns:", merged_df.shape)
```

```python
#STEP 3: Data Cleaning & Feature Engineering

# Create a copy to preserve original dataset
df = merged_df.copy()
```

```python
# Standardize column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
      .str.replace("__", "_")
)

df.columns
```

```python
#Rename Important Columns (Readable Names)
df = df.rename(columns={
    "india/states": "state",
    "nic_name": "industry_name"
})
```

```python
# Extract state name from source_file column
df["state_name"] = (
    df["source_file"]
      .str.replace(".csv", "", regex=False)
      .str.split("_STATE_")
      .str[-1]
      .str.replace("_", " ")
)

df[["source_file", "state_name"]].head()
```

```python
for col in df.columns:
    print(col)
```

```python
# Normalize column names safely
df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(r"[^a-z0-9]+", "_", regex=True)  # replace ANY junk with _
      .str.strip("_")
)

# Verify fixed columns
for col in df.columns:
    print(col)
```

```python
# Workforce aggregation features
df["total_main_workers"] = df["main_workers_total_persons"]
df["total_marginal_workers"] = df["marginal_workers_total_persons"]

df["total_workers"] = (
    df["total_main_workers"] + df["total_marginal_workers"]
)
```

```python
#Gender-Based Workforce Features
# Total male and female workers
df["total_male_workers"] = (
    df["main_workers_total_males"] +
    df["marginal_workers_total_males"]
)

df["total_female_workers"] = (
    df["main_workers_total_females"] +
    df["marginal_workers_total_females"]
)

```

```python
#Rural vs Urban Workforce Features

# Rural workforce
df["total_rural_workers"] = (
    df["main_workers_rural_persons"] +
    df["marginal_workers_rural_persons"]
)
# Urban workforce
df["total_urban_workers"] = (
    df["main_workers_urban_persons"] +
    df["marginal_workers_urban_persons"]
)
```

```python
# FINAL VALIDATION
df[[
    "state_name",
    "industry_name",
    "total_workers",
    "total_male_workers",
    "total_female_workers",
    "total_rural_workers",
    "total_urban_workers"
]].head()
```

```python
#NLP Text Preparation (Industry Names)
# Prepare text for NLP
df["industry_text"] = (
    df["industry_name"]
      .str.lower()
      .str.replace("[^a-z ]", "", regex=True)
      .str.strip()
)
```

```python
# Ensure industry text exists
df["industry_text"] = (
    df["industry_name"]
      .str.lower()
      .str.replace(r"[^a-z ]", "", regex=True)
      .str.strip()
)
```

```python
# Dictionary mapping keywords to industry categories
industry_mapping = {
    "agriculture": ["agriculture", "farming", "crop", "plantation"],
    "poultry": ["poultry", "chicken", "egg", "livestock"],
    "manufacturing": ["manufacturing", "production", "factory", "industrial"],
    "chemical": ["chemical", "fertilizer", "pharmaceutical", "drug"],
    "plastic_rubber": ["plastic", "rubber"],
    "construction": ["construction", "building", "civil"],
    "furniture": ["furniture", "wood"],
    "textile": ["textile", "garment", "apparel", "cloth"],
    "retail": ["retail", "wholesale", "shop", "store", "trade"],
    "transport": ["transport", "logistics", "warehouse"],
    "it_services": ["software", "it", "computer", "information"],
    "others": []
}
```

```python
#NLP Category Assignment Function
def assign_industry_category(text):
    for category, keywords in industry_mapping.items():
        for word in keywords:
            if word in text:
                return category
    return "others"
```

```python
#Apply NLP Categorisation
df["industry_category"] = df["industry_text"].apply(assign_industry_category)
```

```python
import plotly.express as px
```

```python
#State-Wise Total Workforce

state_workforce = (
    df.groupby("state_name", as_index=False)["total_workers"]
      .sum()
      .sort_values(by="total_workers", ascending=False)
)

state_workforce.head()
```

```python
#Bar Chart â€“ Workforce by State
fig_state = px.bar(
    state_workforce,
    x="state_name",
    y="total_workers",
    title="Total Workforce Distribution by State",
)

fig_state.show()
```

```python
#Industry Category Distribution

industry_dist = (
    df.groupby("industry_category", as_index=False)["total_workers"]
      .sum()
      .sort_values(by="total_workers", ascending=False)
)

industry_dist
```

```python
#Pie Chart â€“ Industry Category Share

fig_industry = px.pie(
    industry_dist,
    names="industry_category",
    values="total_workers",
    title="Workforce Share by Industry Category"
)

fig_industry.show()
```

```python
#Gender Distribution Analysis
gender_df = df[["total_male_workers", "total_female_workers"]].sum().reset_index()
gender_df.columns = ["gender", "count"]

gender_df
```

```python
#Gender Distribution Bar Chart

fig_gender = px.bar(
    gender_df,
    x="gender",
    y="count",
    title="Gender-wise Workforce Distribution"
)

fig_gender.show()
```

```python
#Rural vs Urban Workforce Analysis
rural_urban_df = df[["total_rural_workers", "total_urban_workers"]].sum().reset_index()
rural_urban_df.columns = ["area_type", "count"]

rural_urban_df
```

```python
# Rural vs Urban Pie Chart
fig_area = px.pie(
    rural_urban_df,
    names="area_type",
    values="count",
    title="Rural vs Urban Workforce Distribution"
)

fig_area.show()
```

```python
#State Ã— Industry Heatmap
heatmap_df = (
    df.groupby(["state_name", "industry_category"])["total_workers"]
      .sum()
      .reset_index()
)

fig_heatmap = px.density_heatmap(
    heatmap_df,
    x="state_name",
    y="industry_category",
    z="total_workers",
    title="State vs Industry Workforce Heatmap"
)

fig_heatmap.show()
```
